```python
#1. Rock, Paper, Scissors
import random



#Initializes counters for user choices and game results
rkCounter=0
pprCounter=0
scsCounter=0

humanWins=0
cpuWins=0
ties=0

#Takes the user's first input and chooses the computer's first choice randomly
human_choice=raw_input("Let's play Rock, Paper, Scissors!  Enter your choice. ")
computer_choice = random.randint(1,3)  #1 is rock, 2 is paper, 3 is scissors
    
#CPU's "artificially intelligent" algorithm
def cpu_choice(rock,paper,scissors):
    #The first three cases are the obvious cases, where one user choice is the most common and the computer
    #chooses the move that beats that choice
    if rock>paper and rock>scissors:
        choice=2
        print 'CPU chooses paper'
    elif paper>rock and paper>scissors:
        choice=3
        print 'CPU chooses scissors'
    elif scissors>paper and scissors>rock:
        choice=1
        print 'CPU chooses rock.'
        
        #In the following three choices, the computer assumes that the user will pick one of the two
        #equally common choices, and then chooses the choice that beats the other, which will result in
        #either a win or a draw if the human continues his/her trend
    elif rock==paper and rock>scissors:
        choice=2
        print 'CPU chooses paper'
    elif rock==scissors and rock>paper:
        choice=1
        print 'CPU chooses rock.'
    elif paper==scissors and paper>rock:
        choice=3
        print 'CPU chooses scissors'
    #If all three choices are picked equally, the computer chooses randomly
    else:
        choice=random.randint(1,3)
        if choice==1:
            print 'CPU chooses rock.'
        if choice==2:
            print 'CPU chooses paper'
        if choice==3:
            print 'CPU chooses scissors'
            
    return choice
        
    
#This function tests if the human's input is a valid choice
def isValid(choice):
    if human_choice.lower()=='rock' or human_choice.lower()=='paper' or human_choice.lower()=='scissors':
        return True
    else:
        return False
    
#Runs the game until the human chooses quit
while(human_choice.lower()!='quit'):
    
    
    #Testing if the input is rock, paper, or scissors (regardless of case), prompting the user for invalid input
    
    if not (isValid(human_choice)):
        human_choice=raw_input("That's not a choice, please try again. ")
        continue
    
        
    
    #runs the algorithm to choose the best option
    computer_choice=cpu_choice(rkCounter,pprCounter,scsCounter)
    
    #Determines the game result and increments the correct result counter
    if human_choice.lower()=='rock':
        if computer_choice==1:
            print "It's a draw!"
            ties+=1
        elif computer_choice==2:
            print "Sorry, you lost!"
            cpuWins+=1
        else:
            print "Congratulations, you won!"
            humanWins+=1
        rkCounter+=1
    
    elif human_choice.lower()=='paper':
        if computer_choice==1:
            print "Congratulations, you won!"
            humanWins+=1
        elif computer_choice==2:
            print "It's a draw!"
            ties+=1
        else:
            print "Sorry, you lost!"
            cpuWins+=1
        pprCounter+=1
    
    elif human_choice.lower()=='scissors':
        if computer_choice==1:
            print "Sorry, you lost!"
            cpuWins+=1
        elif computer_choice==2:
            print "Congratulations, you won!"
            humanWins+=1
        else:
            print "It's a draw!"
            ties+=1
        scsCounter+=1
        
        
    #Takes the next input
    human_choice=raw_input("Enter your choice. ")
    
#Calculates the total games played
total_games=humanWins+cpuWins+ties

#Prints the results of the game
print 'You played',total_games,'games.  You won',humanWins,'times; you lost',cpuWins,'times; there were',ties,'draws.'
```

```python
#2. Voters in Florida
import re

fp = open("FloridaVoters.html", "r")


#Reads the first line from the file
line=fp.readline()

rows=[]

#This functions reads a single row of the final output table and returns the county name (string),
#Republican voters, and Democrat voters (both ints)
def read_rows(fp,line):
    #The indices are 4 to len-6 because we want to ignore the <td> at the beginning and end
    county=line[4:-6]
    repub=fp.readline()[4:-6]
    #Split and join gets rid of the commas in voter numbers
    comma_split=repub.split(',')
    repub=''.join(comma_split)
    
    
    dem=fp.readline()[4:-6]
    comma_split=dem.split(',')
    dem=''.join(comma_split)
    
    #Reads the three lines that we don't need for our output
    fp.readline()
    fp.readline()
    fp.readline()
    return (county,int(repub),int(dem))


while line is not '':
    #Reads only lines that start with <td>, since those are the rows of the table of interest
    row=re.findall('<td>',line)
    if len(row)>0:
        rows.append(read_rows(fp, line))
    line=fp.readline()
    
    


#Sorts rows by the number of Democratic voters (index 2 of each tuple)
rows.sort(key=lambda x:x[2])

#prints each row
for row in rows:
    print row[0],row[1],row[2]
```

```python
#3. The Google of Quotes
import re
import math



fp = open("quotes.txt","r")
quotes=[]
globalWordBank=[]


quote=fp.readline()
quote=quote.rstrip('\n')

#Part a) Getting a list of quotes

while quote is not '':
    author=fp.readline()
    author=author.rstrip('\n')
    quotes.append(quote+' - '+author)
    
    
    quote=fp.readline()
    
    quote=quote.rstrip('\n')
    
print quotes
```

```python
#Part b)    
def quote_splitter(quote):
    quote=quote.lower() 
    #Strips non-alphanumeric characters from the beginning and end of the strings
    while True:
        if re.match('[\w]',quote[0]):
            index=1
            break
        else:
            quote=quote[1:]
    while True:
        if re.match('[\w]',quote[-1]):
            index=1
            break
        else:
            quote=quote[:-1]
    #Splits the quote by any non-alphanumeric characters: We combined contractions and hyphenated
    #words (e.g. isn't = 'isnt')
    quote=re.sub("['`-]",'',quote)
    word_list=re.split('[^\w]+', quote)
    return word_list
    




split_quotes=[]
#Splits all the quotes into lists of words
for quote in quotes:
    
    split_quotes.append(quote_splitter(quote))
    
    #Creates a global word bank: this will be used for the reverse postings-list dictionary
    for word in quote_splitter(quote):
        if word in globalWordBank:
            continue
        else:
            globalWordBank.append(word)

print split_quotes
```

```python
#Part c)
posting_list_dictionary={}
reverse_pld={}
#Goes through each quote
for index in range(len(split_quotes)):
    posting_list_dictionary[quotes[index]]={}
    count=0
    
    #Goes through each word in each quote
    for word in split_quotes[index]:
        count=0
        #We have to add 1 to the length if the index is 0 or length-1 since there will only be a blank space
        #on one side of the word in these cases.
        if split_quotes[index][0]==word:
            count=count+1
        if split_quotes[index][-1]==word:
            count=count+1
        test_quote=quotes[index].lower()
        test_quote=re.sub("['`\"]",'',test_quote)
        instances=re.findall('[^\w]+'+word+'[^\w]+',test_quote)
        #Counts the number of instances that the word shows up in each quote
        count=count+len(instances)
        posting_list_dictionary[quotes[index]][word]=count
        
print posting_list_dictionary        
```

```python
#Part d)
#Creates the reverse postings-list dictionary by checking every quote to see how many times
#every word occurs in it (using the postings-list dictionary to find the number of occurences)
for word in globalWordBank:
    reverse_pld[word]={}
    for index in range(len(split_quotes)):
        if word in split_quotes[index]:
            reverse_pld[word][quotes[index]]=posting_list_dictionary[quotes[index]][word]
        else:
            continue

print reverse_pld
```

```python
#Part e)
#TF_IDF function
def TF_IDF(word,quote):
    w=posting_list_dictionary[quote][word]
    #gets a list of every key within the quote
    words_list=[]
    for globalWord in globalWordBank:
        if globalWord in quote_splitter(quote):
            words_list.append(posting_list_dictionary[quote][globalWord])
    #Finds the most often used word within the quote        
    q=max(words_list)
    TF=float(w)/q
    
    total_quotes=len(split_quotes)
    occurences=0
    #Countes the number of occurences of the word in every quote
    for q in quotes:
        if word_in_quote(word,q):
            occurences=occurences+1
            
    IDF=math.log(float(total_quotes)/occurences)
    return TF*IDF

#I made this function to determine if a given word is in a quote, since I test this multiple times
#in the program
def word_in_quote(word,quote):
    if word in quote_splitter(quote):
        return True
    else:
        return False

print TF_IDF('augustine', 'Patience is the companion of wisdom. - Saint Augustine')
```

```python
#Part f)            
#Builds a dictionary for a given word with keys that are quotes that contain the word, and values
#that are the TF_IDF scores of each quote
def TFIDF_dictionary_generator(word):
    dictionary={}
    for quote in quotes:
        if word_in_quote(word,quote):
            dictionary[quote]=TF_IDF(word,quote)
            
    return dictionary

print TFIDF_dictionary_generator('augustine')
```

```python
#Part g)
#Builds a dictionary for a given list of words where the keys represent quotes that contain one or more of
#the words and the values represent the sum of TF_IDF scores for each word in the quote
def multiple_word_TFIDF_dictionary_generator(wordList):
    dictionary={}
    for word in wordList:
        for quote in quotes:
            if word_in_quote(word,quote):
                #If the quote is already in the dictionary, then we add the TF_IDF score of the new word
                #(since the quote contains more than one of the given words), otherwise we set it to the
                #TF_IDF score of the first word in that quote
                if quote in dictionary:
                    dictionary[quote]=dictionary[quote]+TF_IDF(word,quote)
                else:
                    dictionary[quote]=TF_IDF(word,quote)
            
    return dictionary

print multiple_word_TFIDF_dictionary_generator(['heart','mind', 'disease'])
```
