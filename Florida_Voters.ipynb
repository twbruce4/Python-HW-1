```python
import re

fp = open("FloridaVoters.html", "r")


#Reads the first line from the file
line=fp.readline()

rows=[]

#This functions reads a single row of the final output table and returns the county name (string),
#Republican voters, and Democrat voters (both ints)
def read_rows(fp,line):
    #The indices are 4 to len-6 because we want to ignore the <td> at the beginning and end
    county=line[4:-6]
    repub=fp.readline()[4:-6]
    #Split and join gets rid of the commas in voter numbers
    comma_split=repub.split(',')
    repub=''.join(comma_split)
    
    
    dem=fp.readline()[4:-6]
    comma_split=dem.split(',')
    dem=''.join(comma_split)
    
    #Reads the three lines that we don't need for our output
    fp.readline()
    fp.readline()
    fp.readline()
    return (county,int(repub),int(dem))


while line is not '':
    #Reads only lines that start with <td>, since those are the rows of the table of interest
    row=re.findall('<td>',line)
    if len(row)>0:
        rows.append(read_rows(fp, line))
    line=fp.readline()
    
    


#Sorts rows by the number of Democratic voters (index 2 of each tuple)
rows.sort(key=lambda x:x[2])

#prints each row
for row in rows:
    print row[0],row[1],row[2]
    
```
